```python
import pandas as pd
```

```python
df.isna().sum()
```

```python
df['Nights'] = df.Nights.fillna(df.Nights.mean())
df['Beds'] = df.Beds.fillna(df.Beds.median())
df['Baths'] = df.Baths.fillna(df.Baths.median())
df['Reviews'] = df.Reviews.fillna(df.Reviews.mean())
```

```python
df.isna().sum()
```

```python
df["Location "] = df['Location '].fillna("Unknown")
```

```python
df.isna().sum()
```

```python
df.City.unique()
```

```python
df['City'] = df.City.replace(
    {
    'Bangalor$' : 'Bangalore',
    'Delhi ' : 'Delhi',
    'Hydrabad' : 'Hyderabad'
    },
regex = True
)
df['City'].unique()
```

```python
df.PPN.unique()
```

```python
df['PPN'] = df.PPN.replace(
    {
    '₹' : '',
    ',' : '',
    'One Thousand Five Hundred' : 1500,
    'Two Thousand' : 2000,
    'Eight Hundred' : 800,
    },
regex = True
).astype(int)
df.head()
```

```python
from dateutil import parser
def format_date(date_str):
        return pd.to_datetime(parser.parse(str(date_str))).strftime('%Y-%m-%d')
df['Recent'] = df['Recent'].apply(format_date)
df.head()
```

```python
df.to_csv('Cleaned.csv' , index = False)
```
